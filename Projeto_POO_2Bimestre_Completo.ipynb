```python
# Desenvolva aqui o c√≥digo da Parte 5
# üìò SISTEMA DE GERENCIAMENTO DE ENQUETES ESCOLARES ‚Äì JHONY (USANDO SQLITE)

import sqlite3

# ----------------------------
# PARTE 1 ‚Äì Classe Principal
# ----------------------------
class Enquete_Jhony:
    def __init__(self, titulo, descricao, ativo):
        self.titulo = titulo
        self.descricao = descricao
        self.ativo = ativo

    def exibir_dados_Jhony(self):
        return f"T√≠tulo: {self.titulo}\nDescri√ß√£o: {self.descricao}\nAtiva: {self.ativo}"

# ----------------------------
# PARTE 2 ‚Äì Heran√ßa e M√©todos
# ----------------------------
class EnqueteComVotos_Jhony(Enquete_Jhony):
    def __init__(self, titulo, descricao, ativo, votos):
        super().__init__(titulo, descricao, ativo)
        self.votos = votos

    def adicionar_voto_Jhony(self):
        self.votos += 1

# ----------------------------
# PARTE 3 ‚Äì Encapsulamento
# ----------------------------
class EnquetePrivada_Jhony:
    def __init__(self, titulo):
        self.__titulo = titulo  # atributo privado

    def get_titulo_Jhony(self):
        return self.__titulo

    def set_titulo_Jhony(self, novo_titulo):
        self.__titulo = novo_titulo

    def titulo_maiusculo_Jhony(self):
        return self.__titulo.upper()

# PARTE 4 ‚Äì Banco de Dados (SQLite)
class EnqueteDB_Jhony:
    def __init__(self, titulo, descricao):
        self.titulo = titulo
        self.descricao = descricao

    def conectar_banco_Jhony(self):
        return sqlite3.connect("enquetes_jhony.db")  # Cria o banco local se n√£o existir

    def salvar_enquete_Jhony(self):
        conn = self.conectar_banco_Jhony()
        cursor = conn.cursor()
        cursor.execute("""
            CREATE TABLE IF NOT EXISTS enquetes (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                titulo TEXT,
                descricao TEXT
            )
        """)
        cursor.execute("INSERT INTO enquetes (titulo, descricao) VALUES (?, ?)",
                       (self.titulo, self.descricao))
        conn.commit()
        conn.close()

    def listar_enquetes_Jhony(self):
        conn = self.conectar_banco_Jhony()
        cursor = conn.cursor()
        cursor.execute("SELECT * FROM enquetes")
        resultados = cursor.fetchall()
        conn.close()
        return resultados

    def excluir_enquete_Jhony(self, id_enquete):
        conn = self.conectar_banco_Jhony()
        cursor = conn.cursor()
        cursor.execute("DELETE FROM enquetes WHERE id = ?", (id_enquete,))
        conn.commit()
        conn.close()

# PARTE 4 ‚Äì Polimorfismo
class EnqueteRapidaDB_Jhony(EnqueteDB_Jhony):
    def mostrar_Jhony(self):
        return f"[R√ÅPIDA] {self.titulo.upper()} - {self.descricao.upper()}"

# PARTE 5 ‚Äì Interface via Terminal (SQLite)
def menu_enquetes_Jhony():
    while True:
        print("\nüìä SISTEMA DE ENQUETES ESCOLARES - JHONY (SQLite)")
        print("1. Criar nova enquete")
        print("2. Listar enquetes")
        print("3. Excluir enquete")
        print("4. Sair")
        opcao = input("Escolha uma op√ß√£o: ")

        if opcao == "1":
            titulo = input("Digite o t√≠tulo da enquete: ")
            descricao = input("Digite a descri√ß√£o da enquete: ")
            enquete = EnqueteDB_Jhony(titulo, descricao)
            enquete.salvar_enquete_Jhony()
            print("‚úÖ Enquete salva com sucesso!")

        elif opcao == "2":
            enquete = EnqueteDB_Jhony("", "")
            enquetes = enquete.listar_enquetes_Jhony()
            if not enquetes:
                print("‚ö†Ô∏è Nenhuma enquete encontrada.")
            else:
                print("\nüìã Lista de Enquetes:")
                for e in enquetes:
                    print(f"ID: {e[0]} | T√≠tulo: {e[1]} | Descri√ß√£o: {e[2]}")

        elif opcao == "3":
            id_excluir = input("Digite o ID da enquete a ser exclu√≠da: ")
            enquete = EnqueteDB_Jhony("", "")
            enquete.excluir_enquete_Jhony(id_excluir)
            print("üóëÔ∏è Enquete exclu√≠da com sucesso.")

        elif opcao == "4":
            print("üëã Saindo do sistema. At√© logo!")
            break

        else:
            print("‚ùå Op√ß√£o inv√°lida. Tente novamente.")

menu_enquetes_Jhony()  
```
